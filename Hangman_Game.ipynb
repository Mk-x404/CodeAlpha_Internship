```python
# HANGMAN GAME

import random

# Predefined words
words = ["cross", "python", "java", "kotlin", "django"]
word = random.choice(words)

# Game state
guessed = ["_"] * len(word)  # to display progress
wrong_guesses = 0  # counter for wrong tries
guessed_letters = []  # track letters already guessed

print("\nWelcome to HANGMAN!")
print("Guess the word, one letter at a time.")
print("You have 6 lives ❤ \n")
print("Type 'exit' anytime to quit the game.\n")

# Main game loop
while wrong_guesses < 6 and "_" in guessed:

    print("-" * 40)
    print("Word:", " ".join(guessed))
    print(f"Wrong guesses: {wrong_guesses}/6")
    
    print(
        f"Guessed letters: {', '.join(guessed_letters) if guessed_letters else 'None'}"
    )

    print(f"Lives left: {'❤' * (6 - wrong_guesses)}")
    print("-" * 40)

    # Ask for input
    guess = input("Guess a letter (or type 'exit' to quit): ").lower()

    # Exit option
    if guess == "exit":
        print("You exited the game. Goodbye!")
        break

    # Validate input
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single valid letter.")
        continue

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    # Check if guess is correct
    if guess in word:
        for i in range(len(word)):
            if word[i] == guess:
                guessed[i] = guess
        print("Good guess!")
    else:
        wrong_guesses += 1
        print("Wrong guess!")

# Game result (only if player didn't exit)
if guess != "exit":
    print("\n" + "=" * 40)
    
    if "_" not in guessed:
        print("Congratulations! You guessed the word:", word)
    
    else:
        print("You lost! The word was:", word)
    print("=" * 40)
```
